```python
class StchasticDescent:
    def __init__(self,lr_rate,epochs):
        self.lr_rate=lr_rate
        self.epochs=epochs
        self.m=None
        
    def fit(self,x_train,y_train):
        self.intercept_=0
        self.coef_=np.zeros(x_train.shape[1])
        n = x_train.shape[0]
        
        for i in range(self.epochs):
            
           for j in range(n):
               idx=np.random.randint(0,x_train.shape[0])
               y_hat=np.dot(x_train[idx],self.coef_)+self.intercept_
               
               coef_der = (2) * np.dot(x_train[idx].T,y_hat - y_train[idx])
               intercept_der = (2)* np.mean(y_hat - y_train[idx])
              
               self.intercept_=self.intercept_-(self.lr_rate* intercept_der)
               self.coef_=self.coef_-(self.lr_rate*coef_der)
            #print(self.intercept_,self.coef_)
        print(self.intercept_,self.coef_)
    def predict (self,x_test):
        return np.dot(x_test,self.coef_)+self.intercept_

sd=StchasticDescent(0.01,610)       
```

```python
class MiniBatchDescent:
    def __init__(self,batch_size,lr_rate,epochs):
        self.lr_rate=lr_rate
        self.epochs=epochs
        self.m=None
        self.batch_size=batch_size
        
    def fit(self,x_train,y_train):
        self.intercept_=0
        self.coef_=np.zeros(x_train.shape[1])
        n = x_train.shape[0]
        
      
        
        for i in range(self.epochs):
            shuffled_indices = np.random.permutation(n)
            x_shuffled = x_train[shuffled_indices]
            y_shuffled = y_train[shuffled_indices]
            
            for j in range(0, n, self.batch_size):
                # Select the mini-batch
                x_batch = x_shuffled[j:j + self.batch_size]
                y_batch = y_shuffled[j:j + self.batch_size]
                
                
                y_hat=np.dot(x_batch,self.coef_)+self.intercept_
                coef_der = (2 / self.batch_size) * np.dot(x_batch.T,y_hat - y_batch)
                intercept_der = (2 / self.batch_size)* np.sum(y_hat - y_batch)
                self.intercept_=self.intercept_-(self.lr_rate* intercept_der)
                self.coef_=self.coef_-(self.lr_rate*coef_der)
            #print(self.intercept_,self.coef_)
        print(self.intercept_,self.coef_)
    def predict (self,x_test):
        return np.dot(x_test,self.coef_)+self.intercept_

mbd=MiniBatchDescent(4,0.1,111)       
```
